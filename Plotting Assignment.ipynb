```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
chilla = pd.read_csv("chillaData.csv")
```

```python
chilla
```

```python
# simple line Grpagh between coding hours and weight of participants

sns.lineplot(x="weight", y="coding_Hours_perday",data=chilla) 
plt.title("Grpagh between coding hours and weight of participants")

plt.show()
```

```python
# lineplot with style( darkgrid)
sns.set_style("darkgrid")
sns.lineplot(x="weight", y="coding_Hours_perday", data=chilla)
plt.title("Grpagh between coding hours and weight of participants (indarkgrid stye)")
plt.show()
```

```python
# lineplot with style( dark)
sns.set_style("dark")
sns.lineplot(x="weight", y="coding_Hours_perday", data=chilla)
plt.title("Grpagh between coding hours and weight of participants (in dark stye)")
plt.show()
```

```python
# lineplot with style(ticks)
sns.set_style("ticks")
sns.lineplot(x="weight", y="coding_Hours_perday", data=chilla)
plt.title("Grpagh between coding hours and weight of participants (in ticks)")
plt.show()
```

```python
# lineplot with style( whitegrid)
sns.set_style("whitegrid")
sns.lineplot(x="weight", y="coding_Hours_perday", data=chilla)
plt.title("Grpagh between coding hours and weight of participants (in whitegrid)")
plt.show()
```

```python
# lineplot with style( whitw)
sns.set_style("white")
sns.lineplot(x="weight", y="coding_Hours_perday", data=chilla)
plt.title("Grpagh between coding hours and weight of participants (in white)")
plt.show()
```

```python
# simple bar Grpagh between field of study and weight of participants

plt.figure(figsize=(10,12))
sns.barplot(x="field_of_study", y="weight", hue = "Gender" ,data= chilla)
plt.title("Grpagh between field of study and weight of participants")
plt.show()
```

```python
# simple bar Grpagh between field of study and weight of participants  ( withoout class interval)
 
plt.figure(figsize=(10,12))
sns.barplot(x="field_of_study", y="weight",hue = "Gender", data= chilla, ci = None)
plt.title("Grpagh between field of study and weight of participants")
plt.show()
```

```python
# simple bar Grpagh between field of study and weight of participants  ( withoout class interval) and saturation
 
plt.figure(figsize=(10,12))
sns.barplot(x="field_of_study", y="weight",hue = "Gender", data= chilla, ci = None,saturation = 0.5)
plt.title("Grpagh between field of study and weight of participants")
plt.show()
```

```python
# simple bar Grpagh between field of study and weight of participants  ( withoout class interval) and saturation
#horizontal
plt.figure(figsize=(10,12))
sns.barplot(x="weight", y="field_of_study",hue = "Gender", data= chilla, ci = None,saturation = 0.5)
plt.title("Grpagh between field of study and weight of participants")
plt.show()
```

```python
# Bar plot between weight and field of study of participants (horizontal)
sns.barplot(x="weight", y="field_of_study", data= chilla,
                       linewidth = 4, facecolor = (1,1,1,0),
                        errcolor="0.2",edgecolor="0.2")
plt.title(" Bar plot between weight and field of study of participants ")
```

```python
# drawoing box plot betweenweight and network of partcipants 
sns.set_style("whitegrid")
sns.boxplot(x="network", y="weight", data= chilla)
plt.show()
```

```python
# box plot with single attribue
sns.set_style("whitegrid")
sns.boxplot(x=chilla["age"])
plt.title("BOx plot of just age column")
plt.show()
```

```python
# drawoing box plot betweenweight and network of partcipants  with dodge true
sns.set_style("whitegrid")
sns.boxplot(x="weight", y="network", data= chilla,dodge = True,hue="Gender")
plt.title("BOx plot of weight and network with dodge true")
plt.show()
```

```python
sns.boxplot(x="weight", y="network", data= chilla,dodge = False,hue="Gender")
plt.title("BOx plot of weight and network with dodge False")
plt.show()
```

```python
# customizing graphs showing customize synbol as *
sns.boxplot(x="network",y="weight",showmeans=True, 
             meanprops = {"marker" : "*",
                          "markersize" : "12",
                           "markeredgecolor" : "blue"},
             data= chilla,)
plt.show()
```

```python
# customizing graphs showing customize synbol as +
sns.boxplot(x="network",y="weight",showmeans=True, 
             meanprops = {"marker" : "+",
                          "markersize" : "12",
                           "markeredgecolor" : "blue"},
             data= chilla,)
plt.show()
```

```python
# customizing graphs showing customize synbol as + and showing label
pt = sns.boxplot(x="network",y="weight",showmeans=True,
             meanprops = {"marker" : "+",
                          "markersize" : "12",               # this is known as dictionary baba g :)
                           "markeredgecolor" : "blue"},
             data= chilla)

plt.title("Customized Box Plot")
pt.set(xlabel='Network of Participants', ylabel='Weight of Participants')   # i have tried a differenrt approach here
plt.show()
```
